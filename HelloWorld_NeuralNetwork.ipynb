```
import tensorflow as tf 
import numpy as np
from tensorflow import keras
```

# Initializing a neural network.

```
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape=[1])])
```

# Compiling the Model

```
model.compile(optimizer='sgd',loss='mean_squared_error')
```

# Preparing the data set. Here we are using numpy arrays for the example.

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0,1.0, 3.0, 5.0, 7.0], dtype=float)
```

# Training the model to fit the data of value one to another (X to Y).

```
model.fit(xs, ys, epochs=500)
```

# Predicting after training the model

```
print(model.predict([10.0]))
```

# A model will never predict the 100% accurate result. We can understand it as  a model is not always positive. So basically it works on probability concepts.
